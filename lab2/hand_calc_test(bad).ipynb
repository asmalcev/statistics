# Семейства вероятностных распределений

```python
import numpy as np
import matplotlib.pyplot as plt
from utils.summary_analysis import analysis, print_analysis


SIZE = 50


dists = [
	np.random.exponential(size=SIZE, scale=0.5),
	np.random.exponential(size=SIZE, scale=1.0),
	np.random.exponential(size=SIZE, scale=1.2),
	np.random.exponential(size=SIZE, scale=1.5),
	np.random.exponential(size=SIZE, scale=2.0),
]
```

## Суммарный анализ

```python
analyzes = list(map(analysis, dists))

for info in analyzes:
	print_analysis(info)
```

## CDF - Функция распределения

```python
def cdf(data, show = True, scalex = 1):
	plt.title('CDF')

	x = np.sort(data) * scalex
	y = 1. * np.arange(len(data)) / (len(data) - 1)

	plt.plot(x, y)

	if show:
		plt.show()

for i in range(0, len(dists)):
	print('Distribution', i + 1)
	cdf(dists[i])
```

### Общий график

Графики распределений были масштабированы так, чтобы их было удобно сравнивать на одном графике 

```python
maxvalues = list(map(max, dists))
maxmaxvalue = max(maxvalues)

for i in range(0, len(dists)):
	cdf(dists[i], False, maxmaxvalue / maxvalues[i])
plt.show()
```

## PDF - Функция плотности вероятности

```python
def pdf(data, show = True):
	plt.title('PDF')

	bins, _ = np.histogram(data, len(data))
	bins = list(map(lambda bin: 1.0 * bin / len(bins), bins))
	plt.plot(bins)

	if show:
		plt.show()

for i in range(0, len(dists)):
	print('Distribution', i + 1)
	pdf(dists[i])
```

## Survivor Function - Функция выжимаемости

```python
def survivor(data, show = True, scalex=1):
	plt.title('Survivor Function')

	x = np.sort(data) * scalex
	y = 1 - np.arange(len(data)) / (len(data) - 1)

	plt.plot(x, y)

	if show:
		plt.show()

for i in range(0, len(dists)):
	print('Distribution', i + 1)
	survivor(dists[i])
```

### Общий график

Графики распределений были масштабированы так, чтобы их было удобно сравнивать на одном графике

```python
for i in range(0, len(dists)):
	survivor(dists[i], False, maxmaxvalue / maxvalues[i])
plt.show()
```

## Log Survivor Function - Логарифм функции выживаемости

```python
def log_survivor(data, show = True):
	plt.title('Log Survivor Function')

	x = np.sort(data)
	y = np.log( 1 - np.arange(len(data)) / (len(data) - 1) )

	plt.plot(x, y)

	if show:
		plt.show()

for i in range(0, len(dists)):
	print('Distribution', i + 1)
	log_survivor(dists[i])
```

## Hazard Function - Функция риска

```python
def hazard(data, show = True):
	plt.title('Hazard Function')

	bins, _ = np.histogram(data, len(data))

	pdf_y = list(map(lambda bin: 1.0 * bin / len(bins), bins))
	survivor_y = 1 - np.arange(len(data)) / (len(data) - 1)

	pdf_y /= survivor_y

	plt.plot(pdf_y)

	if show:
		plt.show()

for i in range(0, len(dists)):
	print('Distribution', i + 1)
	hazard(dists[i])
```
